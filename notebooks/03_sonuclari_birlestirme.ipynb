1- KÜTÜPHANELER VE DOSYA YOLLARI

```python
import pandas as pd
from pathlib import Path

PROJECT_ROOT = Path("..").resolve()

RESULTS_DIR = PROJECT_ROOT / "reports" / "results"
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

KISI1_RESULTS_PATH = RESULTS_DIR / "kisi1_model_sonuclari.csv"
KISI2_RESULTS_PATH = RESULTS_DIR / "kisi2_model_sonuclari.csv"
ALL_RESULTS_PATH = RESULTS_DIR / "tum_model_sonuclari.csv"

print("Proje klasörü:", PROJECT_ROOT)
print("Sonuç klasörü:", RESULTS_DIR)
```

2- KİŞİ1 VE KİŞİ2 SONUÇLARININ OKUNMASI

```python
kisi1_results = pd.read_csv(KISI1_RESULTS_PATH)
kisi2_results = pd.read_csv(KISI2_RESULTS_PATH)

print("Kişi 1 sonuçları:")
display(kisi1_results)

print("Kişi 2 sonuçları:")
display(kisi2_results)
```

3- SÜTUN ADLARININ KONTROLÜ

```python
print("Kişi 1 sütunları:")
print(kisi1_results.columns.tolist())

print("\nKişi 2 sütunları:")
print(kisi2_results.columns.tolist())
```

4- STANDART HÜCRE SUTUN SIRASININ UYGULANMASI

```python
standard_columns = [
    "Model",
    "Accuracy",
    "Weighted Precision",
    "Weighted Recall",
    "Weighted F1-score"
]

kisi1_results = kisi1_results[standard_columns]
kisi2_results = kisi2_results[standard_columns]
```

5- SONUÇLARIN BİRLEŞTİRİLMESİ

```python
all_results = pd.concat(
    [kisi1_results, kisi2_results],
    ignore_index=True
)

print("Birleştirilmiş sonuçlar:")
display(all_results)
```

6- WEİGHTED F1-SCORE'A GÖRE SIRALAMA

```python
all_results_sorted = all_results.sort_values(
    by="Weighted F1-score",
    ascending=False
).reset_index(drop=True)

print("Weighted F1-score değerine göre sıralanmış sonuçlar:")
display(all_results_sorted)
```

7- EN İYİ MODELİN BELİRLENMESİ

```python
best_model = all_results_sorted.iloc[0]

print("En iyi model:")
print("Model:", best_model["Model"])
print("Accuracy:", best_model["Accuracy"])
print("Weighted Precision:", best_model["Weighted Precision"])
print("Weighted Recall:", best_model["Weighted Recall"])
print("Weighted F1-score:", best_model["Weighted F1-score"])
```

8- FİNAL SONUÇLARININ KAYDEDİLMESİ

```python
all_results_sorted.to_csv(
    ALL_RESULTS_PATH,
    index=False,
    encoding="utf-8-sig"
)

print("Tüm model sonuçları kaydedildi:")
print(ALL_RESULTS_PATH)
```

9- KAYDEDİLEN DOSYANIN TEKRAR OKUNMASI VE KONTROL EDİLMESİ

```python
kontrol_df = pd.read_csv(ALL_RESULTS_PATH)

print("Kaydedilen final sonuç tablosu:")
display(kontrol_df)
```

```python
print("Kişi 1 sonuç dosyası var mı?", KISI1_RESULTS_PATH.exists())
print("Kişi 2 sonuç dosyası var mı?", KISI2_RESULTS_PATH.exists())
print("Tüm sonuçlar dosyası var mı?", ALL_RESULTS_PATH.exists())

print("\nModel sayısı:", len(kontrol_df))
print("Model listesi:")
print(kontrol_df["Model"].tolist())
```
